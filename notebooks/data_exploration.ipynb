```python
from datasets import load_from_disk

dataset = load_from_disk("../data/hotpotqa/dev")

print("Total samples:", len(dataset))
```

```python
sample = dataset[0]

print("Question:")
print(sample["question"])

print("\nAnswer:")
print(sample["answer"])

print("\nSupporting facts:")
print(sample["supporting_facts"])

print("\nContext titles (first 5):")
print(sample["context"]["title"][:5])
```

```python
total_samples = len(dataset)

avg_question_len = sum(
    len(s["question"].split()) for s in dataset
) / total_samples

avg_context_docs = sum(
    len(s["context"]["title"]) for s in dataset
) / total_samples

print("Dataset Statistics")
print("------------------")
print(f"Total samples: {total_samples}")
print(f"Average question length: {avg_question_len:.2f} words")
print(f"Average context documents: {avg_context_docs:.2f}")
```

## Observations

- HotpotQA questions are concise and fact-based.
- Each question is linked to multiple context documents.
- Supporting facts identify the relevant documents and sentences.
- The dataset structure supports multi-hop reasoning tasks.
